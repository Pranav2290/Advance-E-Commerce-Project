## **Advance E-Commerce Project**

```python
import pandas as pd

# Orders Data
orders_data = {
    "OrderID": [1,2,3,3,4,5,6,7,8,9,10],
    "CustomerID": [101,102,101,101,103,104,105,106,107,108,109],
    "ProductID": [10,11,12,12,10,13,14,15,11,14,12],
    "OrderDate": ["2025-01-10","2025-01-12","2025-01-15","2025-01-15","2025-02-01",
                  "2025-02-05","2025-02-18","2025-03-03","2025-03-15","2025-03-18","2025-03-20"],
    "Quantity": [1,2,1,1,3,1,2,1,1,3,2],
    "Discount": [5000,2000,200,200,5000,500,100,200,2000,100,200]
}

orders = pd.DataFrame(orders_data)
orders.to_csv("orders.csv", index=False)

# customers Data
customers_data = {
    "CustomerID": [101,102,103,104,105,106,107,108,109],
    "Name": ["Amit","Neha","Sara","John","Reema","Ali","Rita","Karan","Pooja"],
    "City": ["Mumbai","Delhi","Pune","Bangalore","Mumbai","Delhi","Chennai","Pune","Hyderabad"],
    "Age": [29,32,27,None,41,35,29,24,38],
    "Gender": ["M","F","F","M","F","M","F","M","F"],
    "Segment": ["Prime","Regular","Prime","Regular","Prime","Regular","Prime","Regular","Prime"],
    "JoinDate": ["2023-05-01","2022-03-15","2021-06-10",
                 "2020-09-12","2023-01-20","2022-11-10",
                 "2023-07-01","2024-02-18","2021-05-22"]
}

customers = pd.DataFrame(customers_data)
customers.to_csv("customers.csv", index=False)

# Products Data
products_data = {
    "ProductID": [10,11,12,13,14,15],
    "ProductName": ["Laptop","Phone","Shoes","Watch","Headphones","Bag"],
    "Category": ["Electronics","Electronics","Fashion","Fashion","Electronics","Fashion"],
    "Price": [65000,30000,2500,4000,5000,1500],
    "Cost": [50000,25000,1500,2000,3000,800],
    "Rating": [4.5,4.2,3.8,4.0,None,3.5]
}
products = pd.DataFrame(products_data)
products.to_csv("products.csv", index=False)

# Reviews Data
reviews_data = {
    "ReviewID": [1,2,3,4,5,6,7],
    "OrderID": [1,2,3,4,6,8,9],
    "ReviewText": [
        "Excellent laptop, super fast delivery!",
        "Good phone but battery could be better",
        "Nice shoes but size slightly small",
        "Awesome laptop deal, very happy",
        "Watch quality is average, strap loose",
        "Headphones sound good",
        "Phone came late, delivery slow"
    ]
}
reviews = pd.DataFrame(reviews_data)
reviews.to_csv("reviews.csv", index=False)
```

## **Data Cleaning**

### Remove duplicate orders

```python
# there is a duplicate order ID (orderID 3)

# detect duplicated
orders[orders.duplicated(subset = ['OrderID'], keep=False)]

# delete duplicate
orders = orders.drop_duplicates()

# check
orders
```

### Handle missing value

```python
# check for missing value
orders.isnull().sum()
customers.isnull().sum()  # 1 null in age column
products.isnull().sum()   # 1 in rating column
reviews.isnull().sum()
```

```python
# Age replace with median age
customers['Age'].median()

# replace
customers['Age'] = customers['Age'].fillna(customers['Age'].median())


# check mean rating by category
products.groupby('Category')['Rating'].mean() # electronic = 4.35, fashion = 3.76

# Replace missing value with same category mean
products['Rating'] = products.groupby('Category')['Rating'].transform(lambda x: x.fillna(x.mean()))
```

### Add New Business Column  Profit

```python
# To avoid Merge error we now deleting the 'ProfitPerItem' feature
# Note:- add new feature after merging the tables

# 1. first we join all tables then we add profit feature
df = (orders
      .merge(customers, on='CustomerID', how='left')
      .merge(products, on='ProductID', how='left')
      .merge(reviews, on='OrderID', how='left')
      )
df.head(3)
```

```python
# Now we add profit feature
# For that we have to calculate profitPerItem = price - cost
df['ProfitPerItem'] = df['Price'] - df['Cost']

# calculate profi = (profitPerItem * quantity) - discount
df['Profit'] = (df['Quantity'] * df['ProfitPerItem']) - df['Discount']

#check
df.head(3)

```

```python
df.info()
```

Prime customers contribute 4 more profit and purchase more frequently,
but Regular customers provide better profit margins.
This presents two growth opportunities:


1.   onvert Regular customers to Prime,
2.   Optimize discounts for Prime to protect margins.




## **Time Intelligence**

## **Review Text Analysis (NLP Basics)**

```python
# using fillna() replace NaN with empty string in reviewText feature
df['ReviewText'] = df['ReviewText'].fillna("")

# Identify positive and negative keywords
df['PositiveKeyWords'] = df['ReviewText'].str.contains('excellent|awesome|good|happy', case=False)
df['NegativeKeyWords'] = df['ReviewText'].str.contains('bad|slow|late|average', case=False)

df[['OrderID', 'ReviewText', 'PositiveKeyWords', 'NegativeKeyWords']]
```

```python
df['PositiveKeyWords'].sum(),df['NegativeKeyWords'].sum()
```

## **Discount % Analysis (Margin Leakage)**

```python
# 1. Calculate discount percentage (discount/price)*100

# add discountPercent new feature
df['DiscountPercent'] = ((df['Discount'] / df['Price']) * 100).round(2)
```

```python
# 2. Profit Impact by discount

df.groupby('Segment')['DiscountPercent'].sum()
```

```python
import matplotlib.pyplot as plt

# 3. Visualize discount impact by Category
df.groupby('Category')['DiscountPercent'].sum().plot(kind='bar')
plt.ylabel('Discount %')
plt.show()
```
